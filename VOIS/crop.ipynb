```python
def remove_outliers(col_name):
    sorted(col_name)
    Q1 , Q3 = col_name.quantile([0.25, 0.75])
    IQR = Q3 - Q1               #INTER QUANTILE RANGE
    lower = Q1 - 1.5*IQR
    upper= Q3 + 1.5*IQR
    return   lower,upper
```

```python
low,high = remove_outliers(df['K'])
```

```python
low 
```

```python
high

# if value is less than low and higher than high then these values are called outliers
```

```python
# replace the outliers with high value
df['K'] = np.where(df['K']>high,high,df['K'])
```

```python
sns.boxplot(y='K',data=df)
```
